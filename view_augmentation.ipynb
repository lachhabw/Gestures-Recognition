Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

Import the training data

```python
l = pd.read_pickle("./train.pkl")
```

each successive 9 rows are augmentions of the same sample. We will represent the first and the last augmentation of a random row ([-15, -15] version vs [+15, +15] version)

```python
l
```

```python
l_0 = l.iloc[90, :-2].to_numpy().reshape((300, 3))
```

```python
l_1 = l.iloc[98, :-2].to_numpy().reshape((300, 3))
```

```python
l_1.shape
```

```python
l_0.shape
```

### 2 Plot

```python

fig0 = plt.figure(figsize=(20,10))

ax0 = fig0.add_subplot(1, 1, 1, projection="3d")

ax0.set_title("Plot points of all frames together")
ax0.set_xlabel("x")
ax0.set_ylabel("y")
ax0.set_zlabel("z")

ax0.scatter(l_0[:, 0], l_0[:, 1], l_0[:, 2], s=50)


ax0.scatter(l_1[:, 0], l_1[:, 1], l_1[:, 2],color="green" ,s=50)


def animate0(i):

    ax0.view_init(10,i*10)
    return fig0


animation0 = FuncAnimation(fig0, animate0, frames=np.arange(0,36,1), interval=100)
animation0.save("animation123.gif")
```
