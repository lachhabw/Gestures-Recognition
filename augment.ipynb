Augmentation of train set

rotation function


```python
def Rotate_xy(angle_x, angle_y):
    
    anglex_rad = np.deg2rad(angle_x)
    
    cos = np.cos(anglex_rad)
    sin = np.sin(anglex_rad)

    Rx  =  np.array([[1,   0,   0 ],
                     [0, cos, -sin],
                     [0, sin,  cos]]) 



    angley_rad = np.deg2rad(angle_y)

    cos = np.cos(angley_rad)
    sin = np.sin(angley_rad)

    Ry  =  np.array([[cos,  0, sin],
                     [0  ,  1,   0],
                     [-sin, 0, cos]]) 

    return np.dot(Rx, Ry)

```

Augmentation 

```python
for index in range(len(X_train)):

    sample_x = X_train[index, :].reshape((300, 3)); sample_y = y_train[index, :]
    
    for id, angles in enumerate(list(itertools.product([-15, 0, 15], [-15, 0, 15]))):
        
        rotated = np.dot(sample_x, Rotate_xy(angles[0], angles[1]))
        data_augmented[index*9+id, :] = np.hstack((rotated.reshape((900,)), sample_y))
```
